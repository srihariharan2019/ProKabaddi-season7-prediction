```python
from bs4 import BeautifulSoup
from selenium  import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC
import requests
import pandas as pd
import time
driver = webdriver.Chrome()
str1 = '1'
driver.get("https://www.prokabaddi.com/schedule-fixtures-results")
wait(driver, 600).until(EC.presence_of_element_located((By.XPATH, "//*[@class='si-selectBox si-season-change']")))
driver.find_element_by_xpath( "//*[@class='si-selectBox si-season-change']/option[@data-id='1']").click()
driver.implicitly_wait(1)
wait(driver, 600).until(EC.presence_of_element_located((By.XPATH, "//*[@class='si-selectBox si-season-change']")))
driver.implicitly_wait(1)
time.sleep(5)
soup = BeautifulSoup(driver.page_source,'lxml')
name_list = soup.findAll('div', attrs={'class': 'mat-resultStatus'})
driver.implicitly_wait(10)

length1 = len(name_list)
name_list_final = []
for i in range (0,length1):
    name1 = name_list[i].find('span').text
    name_list_final.append(name1)
        
df_result_1 = pd.DataFrame(name_list_final,columns =['match_result'])
df_result_1
driver.close();


        
    
```
